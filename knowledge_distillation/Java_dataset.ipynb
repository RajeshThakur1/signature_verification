```python
def generate_problem_statements(prompt, num_statements):
    response = openai.Completion.create(
        engine="text-davinci-003",  # You can use other engines like "text-davinci-003", "text-davinci-004", etc.
        prompt=prompt,
        max_tokens=150,
        n=num_statements,
        stop=None,
        temperature=0.7,
        frequency_penalty=0.2,
        presence_penalty=0.2
    )
    problem_statements = response['choices'][0]['text'].strip().split('/n')
    return problem_statements
```

```python
def generate_java6_code(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=200,
        stop=None,
        temperature=0.7,
        frequency_penalty=0.2,
        presence_penalty=0.2
    )
    return response['choices'][0]['text'].strip()
```

```python
def generate_java17_code(prompt):
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=200,
        stop=None,
        temperature=0.7,
        frequency_penalty=0.2,
        presence_penalty=0.2
    )
    return response['choices'][0]['text'].strip()
```

```python
prompt = "Generate 20 problem statement to test object-oriented concepts."

```

```python
num_statements = 10
```

```python
problem_statements = generate_problem_statements(prompt, num_statements)
```

```python
ps_List = problem_statements[0].split("\n")
```

```python
for i, statement in enumerate(ps_List):
    print(f"{statement}")
    java6_code = generate_java6_code(statement)
    print("Java 6 Code:")
    print(java6_code)
    print("-" * 50)
    java17_code = generate_java17_code(statement)
    print("Java 17 Code:")
    print(java17_code)
    print("=" * 50)



    print("##########################################################")
```
